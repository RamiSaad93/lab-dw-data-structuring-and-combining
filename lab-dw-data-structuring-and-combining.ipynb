```python
import pandas as pd
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
    
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
df3.columns = ['Customer', 'ST', 'Customer Lifetime Value', 'Education', 'GENDER',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Total Claim Amount', 'Vehicle Class']
```

```python
new_df = pd.concat([df1,df2,df3])
new_df
```

```python
new_df.isna().any()
```

```python
new_df.isna().sum()
```

```python
new_df.dropna(how = "all", inplace = True)
```

```python
new_df.shape
```

```python
new_df.isna().any()
```

```python
new_df.isna().sum()
```

```python
new_df.dtypes
```

```python
new_df["Customer Lifetime Value"] = new_df["Customer Lifetime Value"].apply(lambda x: x.strip("%") if isinstance(x,str) else x)
```

```python
new_df
```

```python
new_df.dtypes
```

```python
new_df["Customer Lifetime Value"] = new_df["Customer Lifetime Value"].apply(float)
```

```python
new_df["Number of Open Complaints"] = new_df["Number of Open Complaints"].apply(lambda x: x.split("/")[1] if isinstance(x,str) else x)
```

```python
#new_df["Number of Open Complaints"] = new_df["Number of Open Complaints"].str.split("/").str[1]
```

```python
new_df["Number of Open Complaints"] = new_df["Number of Open Complaints"].apply(int) 
```

```python
new_df.dtypes
```

```python
new_df
```

```python
new_df['Customer Lifetime Value'].fillna(new_df['Customer Lifetime Value'].mean(), inplace=True)
```

```python
new_df.isna().sum()
```

```python
new_df.GENDER.value_counts()
```

```python
gender_dict = {"Male":"M", "female":"F", "Femal":"F", "F":"F", "M":"M"}
new_df.GENDER = new_df.GENDER.map(gender_dict)

new_df.GENDER.value_counts()
```

```python
new_df.GENDER.isna()
```

```python
new_df.GENDER.fillna(method = "bfill", inplace = True)
```
